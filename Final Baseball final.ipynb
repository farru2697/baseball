# Data Preprocessing

```python
X=new_df.drop("W",axis=1)
y=new_df["W"]
```

```python
X
```

```python
y
```

```python
X.value_counts()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import StandardScaler

stdscale = StandardScaler().fit(X_train)
x_train_std = stdscale.transform(X_train)
x_train_std = stdscale.transform(X_test)
```

```python
y_train       
```

# MultiLinear Regression

```python
# Fitting Linear Regression to the Training dataset
from sklearn.linear_model import LinearRegression

LM = LinearRegression()
LM.fit(X_train, y_train)
```

```python
# predicting the Test set result
y_pred = LM.predict(X_test)
```

```python
# calculating the coefficients
LM.coef_    #m1, m2 , m3 value
```

```python
# calculating the Intercept
LM.intercept_
```

```python
# Calculating the R squared value
from sklearn.metrics import r2_score
R2sq = r2_score(y_test, y_pred)


from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)


import math
rmse = math.sqrt(MSE)
rmse

from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, y_pred)

from sklearn.metrics import mean_absolute_percentage_error
MAPE = mean_absolute_percentage_error(y_test, y_pred)

from sklearn.metrics import mean_squared_log_error
MSLE = mean_squared_log_error(y_test, y_pred)

from sklearn.metrics import max_error
Maxerror = max_error(y_test, y_pred)


from sklearn.metrics import explained_variance_score
EVS = explained_variance_score(y_test, y_pred)

print('R2sq: ', round(R2sq,4))
print('MSE: ', round(MSE,4))
print('rmse: ', round(rmse,4))
print('MAE: ', round(MAE,4))

print('MAPE : ', round(MAPE ,4))
print('MSLE: ', round(MSLE,4))
print('Maxerror: ', round(Maxerror,4))
print('EVS: ', round(EVS,4))
```

```python
from sklearn.model_selection import cross_val_score

print(cross_val_score(LM,X,y,cv=5).mean())
```

```python
LM.score(X_train , y_train)
```
